```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
happiness = pd.read_csv("happiness_2019.csv")
```

```python
happiness = happiness.sort_values(by=["Country or region"])
happiness
```

```python
# happiness_two = happiness
happiness_html = happiness.to_html(index=False).replace('\n', '')
# happiness_html= list(map(lambda x:x.strip("\n"),happiness_html))
happiness_html
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Happpiness Scores by Country')
ax1.set_ylabel('Happiness Index')
ax1.set_xlabel('Range')
ax1.boxplot(happiness["Score"])
plt.show();
fig1.savefig('assets/images/boxplot.jpeg')
```

```python
plt.scatter(happiness["Country or region"], happiness["Score"], marker="o", facecolor="blue" )
```

```python
# Happiness by Generosity Level

plt.scatter(happiness["Generosity"], happiness["Score"], marker="o", c=happiness["Overall rank"])
plt.title("Happiness by Generosity")
plt.ylabel("Happiness Index")
plt.xlabel("Generosity Index")
plt.colorbar();
plt.savefig('Resources/assets/images/generosity.png')
```

```python
# Happiness by freedoms and civil liberties

plt.scatter(happiness["Freedom to make life choices"], happiness["Score"], marker="o", c=happiness["Overall rank"])
plt.title("Happiness by Autonomy")
plt.ylabel("Happiness Index")
plt.xlabel("Autonomy Index")
plt.colorbar();
plt.savefig('Resources/assets/images/autonomy.png')
```
